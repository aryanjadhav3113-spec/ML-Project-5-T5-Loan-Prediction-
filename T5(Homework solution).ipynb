```python
# for dataframe operations
import pandas as pd

#data visulaziations
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# lets import the dataset using the read_csv function
data = pd.read_csv(r"C:\Users\Aryan\OneDrive\Desktop\LoanData.csv")

#Lets check the shape of the dataset
data.shape
```

Make a Heat Map

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['figure.figsize']=(5,5)
plt.style.use('fivethirtyeight')

# select only numeric columns
numeric_data = data.select_dtypes(include=['number'])

sns.heatmap(numeric_data.corr(),
            annot=True,
            fmt='.2f',
            cmap='Wistia'

)
plt.title('Correlation Heatmap')
plt.show()
```

Compare Loan Amount and Applicant Income

```python
sns.scatterplot(x=data['ApplicantIncome'], y=data['LoanAmount'])

plt.title('Applicant Income vs Loan Amount')
plt.show()
```
